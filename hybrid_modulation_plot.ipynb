```python
from matplotlib import pyplot as plt
from matplotlib import cm
import numpy as np
import itertools
%matplotlib widget
```

```python
fig, ax = plt.subplots(figsize=(12, 12))

n = 10
plot_count = 150

for full_scale_ratio, c in zip(np.linspace(0.01, 0.9, plot_count), np.linspace(0, 1, plot_count)):
    steps_high = np.logspace(0, n-1, num=n, base=2)
    steps_low = steps_high * full_scale_ratio
    sel = np.vstack([np.zeros(n), steps_low, steps_high]).swapaxes(0, 1)
    index = np.array(list(itertools.product(range(3), repeat=n)))
    collected = np.take_along_axis(sel[np.newaxis, :, :], index[:,:,np.newaxis], axis=2)
    all_values = sorted(collected[:,:,0].sum(axis=1))
    #ax.scatter(all_values, all_values)
    ax.plot(all_values, color=cm.gray(c), lw=.5)
    
    
```
